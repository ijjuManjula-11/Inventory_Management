```python
#Accessing data in Table format 
def print_items_table(items):
    print(f"{'Item ID':<8} {'Item Name':<12} {'Category':<15} {'Quantity':<10} {'Price/Unit':<12}")
    print("-" * 60)

    for item in items:
        print(f"{item[0]:<8} {item[1]:<12} {item[2]:<15} {item[3]:<10} {item[4]:<12}")
```

```python
print_items_table(items)
```

```python
#Add an item 
def add_item(items):
    print("Enter new item details:")

    item_id = int(input("Item ID: "))
    
    # Check if ID already exists
    for item in items:
        if item[0] == item_id:
            print("Item ID already exists! Cannot add.")
            return

    item_name = input("Item Name: ")
    category = input("Category: ")
    quantity = int(input("Quantity: "))
    price = float(input("Price per Unit: "))

    # Append new item to the list
    items.append([item_id, item_name, category, quantity, price])
    print("\nItem added successfully!")



```

```python
add_item(items)
```

```python
def search_an_item(items, search_id):
    for item in items:
        if item[0] == search_id:
            return item
    return None
```

```python
def search_item_name(items,item_name):
    result=[]
    for item in items:
        if item[1].lower() == item_name.lower():
            result.append(item)
    return result
```

```python
def delete_item(items, item_id):
    # Find the item index
    for index, item in enumerate(items):
        if item[0] == item_id:
            items.pop(index)
            print(f"Item ID {item_id} deleted successfully!")
            return
    print("Item not found!")
```

```python
def sales_report(items):
    report = {}

    for item in items:
        item_name = item[1]
        quantity = item[3]

        if item_name in report:
            report[item_name] += quantity
        else:
            report[item_name] = quantity

    print("\nSALES REPORT (Total Quantity Sold)")
    print(f"{'Item Name':<15} {'Total Quantity':<15}")
    print("-" * 30)

    for name, qty in report.items():
        print(f"{name:<15} {qty:<15}")

```

```python
sales_report(items)
```

```python
#exporting the data to CSv file 
import csv
def write_items_to_csv(items, filename="items.csv"):
    with open(filename, "w", newline="") as file:
        writer = csv.writer(file)

        # Write header
        writer.writerow(["Item ID", "Item Name", "Category", "Quantity", "Price per Unit"])

        # Write item rows
        for item in items:
            writer.writerow(item)

    print("Data written to CSV successfully!")
```

```python
write_items_to_csv(items,filename="items.csv")
```

```python
def check_unique_item_ids(items):
    item_ids = [item[0] for item in items]

    if len(item_ids) == len(set(item_ids)):
        print("All Item IDs are unique ")
    else:
        print("Duplicate Item IDs found ")
```

```python
item_category_map = {
    "Laptop": "Electronics",
    "Notebook": "Stationery",
    "Pen": "Stationery",
    "Chair": "Furniture",
    "Table": "Furniture",
    "Mouse": "Electronics",
    "Keyboard": "Electronics",
    "Monitor": "Electronics",
    "Bottle": "Groceries",
    "Bag": "Accessories"
}
```

```python
def update_item_name(items, item_id, new_name):
    for item in items:
        if item[0] == item_id:
            item[1] = new_name
            item[2] = item_category_map.get(new_name, "Unknown")#notebook 
            print("Item updated successfully")
            print(item)
            return
    print("Item not found")
update_item_name(items,102,"Notebook")
```
